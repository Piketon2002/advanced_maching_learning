```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('14_25.csv', header=None)
```

```python
df
```

```python
# будет 3 главные компоненты
pca = PCA(n_components=3, svd_solver='full')
pca.fit(df)
df_new = pca.transform(df) # новые координаты объектов (получается у каждого объекта теперь только 3 координаты, а не 10)
```

```python
# посмотреть компоненты векторов компонент (это по сути вектора phi и в итоге исх матрица умножить на phi_i это будет
# столбец i-ых координат всех объектов)
pca.components_
```

```python
# доля объясненной дисперсии для каждой компоненты
print(sum(pca.explained_variance_ratio_))
```

```python
pca = PCA(n_components=2, svd_solver='full')
pca.fit(df)
df_new = pca.transform(df)
```

```python
# Какое количество групп объектов можно выделить, если использовать только первые две главных компоненты?
first_comp = [pair[0] for pair in df_new]
second_comp = [pair[1] for pair in df_new]
plt.scatter(first_comp, second_comp)
plt.show()
```
