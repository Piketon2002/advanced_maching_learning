```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df_ves = pd.read_csv('X_loadings_648.csv', header=None)
df_schet = pd.read_csv('X_reduced_648.csv', header=None)
```

```python
df_schet
```

```python
def get_matrix(df):
    ans = [df.loc[i,0].split(';') for i in range(len(df))]
    correct_ans = []
    for item in ans:
        temp = [float(num) for num in item]
        correct_ans.append(temp)
    return correct_ans    
```

```python
df_weights = np.asarray(get_matrix(df_ves))
df_vectors = np.asarray(get_matrix(df_schet))

# транспонирование
vect_trans = np.transpose(df_weights)
```

```python
res = np.matmul(df_vectors,vect_trans)
```

```python
res.shape
```

```python
fig, ax = plt.subplots()
im = ax.imshow(res, cmap='viridis')

cbar = ax.figure.colorbar(im, ax=ax)
plt.show()
```
